```python
#menampilkan data yang memiliki pelaporan tindakan kriminal sepanjang tahun 2008-2016 di London 
criminal = df[df['value'] >= 1]
criminal
```

```python
#mencari jumlah laporan kriminal berdasarkan masing-masing major category 
df['major_category'].value_counts()
```

```python
#mencari jumlah laporan kriminal berdasarkan masing-masing minor category 
criminal['minor_category'].value_counts()
```

```python
#menampilkan jumlah kriminalitas yang di grouping berdasarkan tahun dan major_category
count_crimeperyear = criminal.groupby(['year','major_category']).count()['value'].reset_index()
count_crimeperyear
```

```python
#mencari nilai jumlah kejahatan yang dilaporkan berdasarkan wilayah dan tahun
sum_crimeperyear = criminal.groupby(['year','major_category']).sum()['value'].reset_index()
sum_crimeperyear
```

```python
#mencari nilai jumlah kejahatan yang dilaporkan berdasarkan wilayah dan tahun
sum_crimeperyear = criminal.groupby(['year','borough']).sum()['value'].reset_index()
sum_crimeperyear
```

```python
#menampilkan data laporan kriminal berdasarkan wilayah di masing-masing tahun 2008-2016
borough_sum = sum_crimeperyear.reset_index().pivot(columns = 'year', index = 'borough', values= 'value').reset_index()
borough_sum = borough_sum.replace(np.nan,0)
borough_sum
```

```python
theft = criminal[criminal["major_category"].isin(["Theft and Handling"])]
theft
```

```python
sum_theftperyear = theft.groupby(['year','borough']).sum()['value'].reset_index()
sum_theftperyear
```

```python
theft_sum = sum_theftperyear.pivot(columns = 'year', index = 'borough', values= 'value')
theft_sum = theft_sum.replace(np.nan,0)
theft_sum
```

```python
theft_total = theft_sum.sum(axis=1).reset_index()
theft_total
```

```python
#data jumlah kriminalitas berdasarkan major category sepanjang tahun 2008-2016
sum_crimeperyear1 = sum_crimeperyear.pivot(columns = 'year', index = 'major_category', values= 'value')
sum_crimeperyear1 = sum_crimeperyear1.replace(np.nan,0)
sum_crimeperyear1 = sum_crimeperyear1.T
sum_crimeperyear1
```

```python
#menampilkan grafis area plot perkembangan kriminalitas  
sum_crimeperyear1.plot(kind='area', alpha=0.5, figsize=(20,10))
plt.title('Area Plot of Crime based on Major Category') # add a title to the area plot
plt.ylabel('Number of Crime') # add y-label
plt.xlabel('Year') # add x-label
plt.show()
```

grafik menunjukkan angka kriminalitas tertinggi yang dilaporkan sepanjang tahun 2008-2016 di London adalah Kategori Theft & Handling. untuk kasus berdasarkan Violence Against the person mengalami peningkatan di 2016

```python
#menampilkan jumlah laporan kriminal berdasarkan major category
majoritycat_count = criminal['major_category'].value_counts()
majoritycat_count
```

```python
#membuar grafik bar
criminal['major_category'].value_counts().plot(kind='barh',figsize=(20,3))
for i, cat in enumerate(majoritycat_count) :
 plt.text(x=cat, y=i-0.2, s=cat)
plt.title('Area Plot of Crime based on Major Category') # add a title to the area plot
plt.ylabel('Major Category') # add y-label
plt.xlabel('NUmber of Crime') # add x-label
plt.show()
```

```python
criminal['major_category'].value_counts().plot(kind='bar',figsize=(15,8))
for i, cat in enumerate(majoritycat_count) :
    plt.text(x=i-0.25, y=cat, s=cat)
plt.title('Area Plot of Crime based on Major Category') # add a title to the area plot
plt.ylabel('Major Category') # add y-label
plt.xlabel('NUmber of Crime') # add x-label
plt.show()
```

Grafik Bar menunjukkan tindak kriminal yang dilaporkan di London pada Tahun 2008-2016 tertinggi adalah kategori theft and handling sebanyak 1136994 kasus, dan terendah disebabkan sexual offences sebanyak 1075 kasus.

```python
criminal['major_category'].value_counts().plot(kind='pie', autopct='%.2f%%', labels=None, legend=True, figsize=(20,10))
plt.legend()
plt.show()
```

proporsi terbesar angka kriminalitas di London tahun 2008-2016 disebabkan oleh Theft and Handling, sebanyak 33.25% dari total laporan kasus kriminalitas
